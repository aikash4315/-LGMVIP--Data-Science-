```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
vk=pd.read_csv('terriosm.csv')
```

```python
vk.head()
```

```python
vk.tail()
```

```python
vk=vk[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
       "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
       "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python
vk.head()
```

```python
vk.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
vk.head()
```

```python
vk.describe()
```

```python
vk.isnull().sum()
```

```python
vk["Killed"]=vk["Killed"].fillna(0)
vk["Wounded"]=vk["Wounded"].fillna(0)
vk["Casualty"]=vk["Killed"]+vk["Wounded"]
```

```python
vk.isnull().sum()
```

```python
attacks=vk["Year"].unique()
attacks_count=vk['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(20,10))
sns.barplot(x=attacks,y=attacks_count,palette="tab10" )
plt.xticks(rotation=50)
plt.xlabel('attacking years',fontsize=20)
plt.ylabel('number of attacks',fontsize=20)
plt.show()
```

```python
yk=vk[["Year","Killed"]].groupby("Year").sum()
yk.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)



yk.plot(kind="bar",color="cornflowerblue",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed")
```

```python
reg=pd.crosstab(vk.Year,vk.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

```python
rc=vk[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

```python
ct=vk["Country"].value_counts().head(10)
ct
```

```python
ct.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Country wise Attacks",fontsize=13)
plt.xlabel("Countries",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
ck=vk[["City","Killed"]].groupby("City").sum().sort_values(by="Killed",ascending=False).drop("Unknown")
ck.head(10)
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ck[:10].plot(kind="bar",color="cornflowerblue",figsize=(15,6),ax=ax0)
ax0.set_title("People Killed in each City")
ax0.set_xlabel("Cities")
ax0.set_ylabel("Number of People Killed")
```

```python
ta=vk["Target Type"].value_counts()
ta
```

```python
ta.plot(kind="bar",color="cornflowerblue",figsize=(15,6))
plt.title("Types of Targets",fontsize=13)
plt.xlabel("Target Types",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()
```

```python
tk=vk[["Target Type","Killed"]].groupby("Target Type").sum().sort_values(by="Killed",ascending=False)
tk
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
tk.plot(kind="bar",color="cornflowerblue",figsize=(30,8),ax=ax0)
ax0.set_title("People Killed in each Target Attack")
ax0.set_xlabel("Target Types")
ax0.set_ylabel("Number of people Killed")
```
